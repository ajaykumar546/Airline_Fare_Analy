```python
file.drop(['City1'], axis = 1, inplace = True)
file.drop(['City2'], axis = 1, inplace = True)
file.drop(['Market Leading Airline (MLA)'], axis = 1, inplace = True)
file.drop(['Low Price Airline (LPA)'], axis = 1, inplace = True)
```

### 4.2.1. Rescaling the Features

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
# Apply scaler() to all the columns except the 'dummy' variables
vars = ['Average Flight Fare', 'Distance', 'Average Weekly Passengers', 'Market Share MLA', 'Average MLA Fare'
            ,'Market Share LPA', 'Average LPA Fare']

file[vars] = scaler.fit_transform(file[vars])
```

```python
file.head()
```

```python
file.describe().transpose()
```

###  4.2.2. Splitting Data into Training and Test

```python
from sklearn.model_selection import train_test_split
np.random.seed(0)
file_train, file_test = train_test_split(file, train_size = 0.7, test_size = 0.3, random_state = 100)
```

```python
y_train = file_train.pop('Average Flight Fare')
X_train = file_train
```

```python
y_test = file_test.pop('Average Flight Fare')
X_test = file_test
```

```python
X_test
```

```python
file.shape
```

```python
X_train.shape
```

### 4.2.3. Design and Fit a Neural Network

```python
import tensorflow
```

```python
from tensorflow.keras.models import Sequential
model = Sequential()
model.add(tensorflow.keras.layers.BatchNormalization(input_shape=(6,)))
model.add(tensorflow.keras.layers.Dense(1))
```

```python
# Compile the model
model.compile(loss=tensorflow.keras.losses.MeanSquaredError(), optimizer="sgd")

# Fit the model
model.fit(x=X_train, y=y_train, epochs=10, validation_data=(X_test, y_test))
```

### evaluate the model

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score, accuracy_score
```

```python
mse = mean_squared_error(y_test, y_pred)
mse
```

```python
mean_absolute_error(y_test, y_pred)
```

```python
y_train_predict=model.predict(X_train)
print('r2_score',r2_score(y_train,y_train_predict))
```
